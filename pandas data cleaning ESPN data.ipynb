# Renaming the columns for better understanding

```python
df = df.rename(columns = {'Mat': 'Matches', 'NO': 'Not_Outs', 'HS': 'Highest_inn_score', 'BF': 'Balls_Faced', 'SR': "Strike_Rate"})
```

## Checking for null values if any

```python
df.isnull().any()
```

## Filling the values with 0 as null represents directly the balls faced and the stricke rate

```python
df['Balls_Faced'] = df['Balls_Faced'].fillna(0)
```

```python
df['Strike_Rate'] = df['Strike_Rate'].fillna(0)
```

## Checking for duplicate values

```python
df[df['Player'].duplicated()==1]
```

```python
df
```

## Splitting the Span column into two columns i.e., Rookie_Year and Final_year

```python
df['Span'].str.split(pat = '-')
```

```python
df['Rookie_Year'] = df['Span'].str.split(pat = '-').str[0]
```

```python
df
```

```python
df['Final_year'] = df['Span'].str.split(pat = '-').str[1]
df
```

## Now dropping Span column as it is not necessary anymore

```python
df.drop(['Span'], axis = 1, inplace = True)
```

```python
df
```

## Removing country codes under the player names and adding separate columns for those

```python
df['Country'] = df['Player'].str.split(pat = '(').str[1]
df
```

```python
df['Country'] = df['Country'].str.split(pat = ')').str[0]
df
```

```python
df['Player'] = df['Player'].str.split(pat = '(').str[0]
df
```

```python
df.info()
```

## Changing the data types of the column

```python
df['Highest_inn_score'] = df['Highest_inn_score'].str.split(pat = '*').str[0]
df
```

```python
df.info()
```

```python
df['Highest_inn_score'] = df['Highest_inn_score'].astype(int)
```

```python
df.info()
```

```python
df = df.astype({'Rookie_Year': 'int', 'Final_year': 'int'})
```

```python
df.info()
```
